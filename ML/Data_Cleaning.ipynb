```python
#Clean up the pollutant data
pollution_df = pd.read_csv(r'resources/us_pollution_2000_2016.csv')
display(pollution_df)
```

```python
#drop unneeded columns, EPA focuses on the highest level of pollutant reported and measures aqi based on county so I adjusted the dataframe to reflect this
pdf1= pollution_df.drop(columns=['city', 'no2avg', 'no2maxhour', 'no2aqi', 'o3avg', 'o3maxhour', 'o3aqi', 'so2avg', 'so2maxhour', 'so2aqi', 'coavg', 'comaxhour', 'coaqi'])
display(pdf1)
```

```python
#drop extra date rows, since I am only focusing on the highest value from each day
pdf2= pdf1.drop_duplicates(subset=['usstate','county','local_date'])
display(pdf2)
```

```python
pdf3= pdf2.drop(pdf2.columns[0], axis=1)
display(pdf3)
```

```python
pdf2.to_csv(filepath2, index=False)
```
