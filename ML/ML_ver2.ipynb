```python
#Clean up the pollutant data
pollution_df = pd.read_csv(r'resources/us_pollution_2000_2016.csv')
display(pollution_df)
```

```python
#drop unneeded columns, EPA focuses on the highest level of pollutant reported so I adjusted the dataframe to reflect this
pdf1= pollution_df.drop(columns=['no2avg', 'no2maxhour', 'no2aqi', 'o3avg', 'o3maxhour', 'o3aqi', 'so2avg', 'so2maxhour', 'so2aqi', 'coavg', 'comaxhour', 'coaqi'])
display(pdf1)
```

```python
#drop extra date rows, since I am only focusing on the highest value from each day
pdf2= pdf1.drop_duplicates(subset=['usstate', 'county', 'city', 'local_date'])
display(pdf2)
```

```python
#drop rows not in a city
pdf3= pdf2[pdf2.city !='Not in a city']
display(pdf3)
```

```python
merged_df=pd.merge(pdf3, aqi_df1, on=['usstate', 'county', 'local_date'], how='inner')
display(merged_df)
```

```python
#drop string columns and convert local_date to integer to prep for machine learning model
la_ml_df1 = la_df.drop(columns=['usstate', 'county', 'city'])
display(la_ml_df1)
```

```python
from pandas.plotting import scatter_matrix
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
```

```python
#Separate clean dataframe into features and target variable 
X = la_ml_df2[['no2maxvalue', 'o3maxvalue', 'so2maxvalue', 'comaxvalue']].values
y = la_ml_df2['AQI'].values
```

```python
#normalize or scale(using min max scaler) data
scaler = MinMaxScaler()
scaled = scaler.fit_transform(X,y)
display(scaled)
```

```python
#Separate clean dataframe into features and target variable 
X = scaled[['no2maxvalue', 'o3maxvalue', 'so2maxvalue', 'comaxvalue']].values
y = scaled['AQI'].values
```

```python
#Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#Linear regression model 
from sklearn.linear_model import LinearRegression
lin_regr = LinearRegression()
lin_regr.fit(X_train, y_train)
```

```python
y_pred = lin_regr.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
pred_y_df = pd.DataFrame({'Actual Value':y_test, 'Predicted Value':y_pred, 'Difference': y_test-y_pred})
display(pred_y_df)
```
