```python
import pandas as pd
import numpy as np
```

```python
pollution_file = 'us_pollution.csv'
pollution_data = pd.read_csv(pollution_file, low_memory = False)
```

```python
us_pollution_df = pd.DataFrame(pollution_data)
```

```python
us_pollution_df.head()
```

```python
us_pollution_df.columns
```

```python
us_pollution_df = us_pollution_df[['State', 'County', 'City', 'Date Local', 
                                  'NO2 Mean','NO2 1st Max Value', 'NO2 1st Max Hour', 'NO2 AQI',
                                  'O3 Mean', 'O3 1st Max Value', 'O3 1st Max Hour', 'O3 AQI', 
                                  'SO2 Mean', 'SO2 1st Max Value', 'SO2 1st Max Hour', 'SO2 AQI',
                                  'CO Mean', 'CO 1st Max Value', 'CO 1st Max Hour', 'CO AQI']]
us_pollution_df.head()
```

```python
def change_column_name(old_name, new_name):
    if old_name in us_pollution_df:
        us_pollution_df[new_name] = us_pollution_df.pop(old_name)
change_column_name('State', 'usstate')
change_column_name('County', 'county')
change_column_name('City', 'city')
change_column_name('Date Local', 'local_date')

change_column_name('NO2 Mean', 'no2avg')
change_column_name('NO2 1st Max Value', 'no2maxvalue')
change_column_name('NO2 1st Max Hour', 'no2maxhour')
change_column_name('NO2 AQI', 'no2aqi')

change_column_name('O3 Mean', 'o3avg')
change_column_name('O3 1st Max Value', 'o3maxvalue')
change_column_name('O3 1st Max Hour', 'o3maxhour')
change_column_name('O3 AQI', 'o3aqi')

change_column_name('SO2 Mean', 'so2avg')
change_column_name('SO2 1st Max Value', 'so2maxvalue')
change_column_name('SO2 1st Max Hour', 'so2maxhour')
change_column_name('SO2 AQI', 'so2aqi')

change_column_name('CO Mean', 'coavg')
change_column_name('CO 1st Max Value', 'comaxvalue')
change_column_name('CO 1st Max Hour', 'comaxhour')
change_column_name('CO AQI', 'coaqi')
us_pollution_df.head()
```

```python
us_pollution_df = us_pollution_df[us_pollution_df['coaqi'].notna()]
us_pollution_df = us_pollution_df[us_pollution_df['so2aqi'].notna()]
us_pollution_df['local_date'] = pd.to_datetime(us_pollution_df['local_date'])
us_pollution_df.head()
```

```python
us_pollution_df.to_csv('us_pollution_2000_2016.csv', index=False)
```

```python
aqi2022_file = 'annual_aqi_by_cbsa_2022.csv'
aqi2022_data = pd.read_csv(aqi2022_file, low_memory = False)
combined_df = pd.DataFrame(aqi2022_data)
```

```python
def combine(file, df):
    data = pd.read_csv(file, low_memory = False)
    new_df = pd.DataFrame(data)
    df = pd.concat([new_df, df], axis=0)
    return df
```

```python
for year in range(22):
    if year < 10:
        file = f'annual_aqi_by_cbsa_200{year}.csv'
        combined_df = combine(file,combined_df)
    else:
        file = f'annual_aqi_by_cbsa_20{year}.csv'
        combined_df = combine(file,combined_df)
```
