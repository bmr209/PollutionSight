```python
import pandas as pd
import numpy as np
```

```python
pollution_file = 'us_pollution_2000_2016.csv'
pollution_data = pd.read_csv(pollution_file, low_memory = False)
census_file = 'census_data.csv'
census_data = pd.read_csv(census_file, low_memory = False)
```

```python
us_pollution_df = pd.DataFrame(pollution_data)
census_df = pd.DataFrame(census_data)
```

```python
census_df["usstate"] = census_df["usstate"].str.lower()
census_df["city"] = census_df["city"].str.lower()
census_df.head()
```

```python
us_pollution_df["usstate"] = us_pollution_df["usstate"].str.lower()
us_pollution_df["county"] = us_pollution_df["county"].str.lower()
us_pollution_df["city"] = us_pollution_df["city"].str.lower()
us_pollution_df.head()
```

```python
index = 0
state = us_pollution_df.loc[index,"usstate"]
city = us_pollution_df.loc[index,"city"]
year = us_pollution_df.loc[index,"local_year"]
population_year = f"POPESTIMATE{year}"

population = census_df[population_year].loc[(census_df["usstate"] == 'arizona') & (census_df["city"] == 'phoenix city')].values[0]
us_pollution_df.loc[index,"population"] = population 
us_pollution_df.head()
```

```python
for index, row in us_pollution_df.iterrows():
    state = us_pollution_df.loc[index,"usstate"]
    city = us_pollution_df.loc[index,"city"]
    year = us_pollution_df.loc[index,"local_year"]
    population_year = f"POPESTIMATE{year}"
    
    try:
        population = census_df[population_year].loc[(census_df["usstate"] == state) & (census_df["city"] == city)].values[0]
        us_pollution_df.loc[index,"population"] = population 
        print(f"It was a town {index}")
    except:
        try:
            city = f"{city} city"
            population = census_df[population_year].loc[(census_df["usstate"] == state) & (census_df["city"] == city)].values[0]
            us_pollution_df.loc[index,"population"] = population 
            print(f"wops it was a city {index}")
        except:
            try:
                city = f"{city} township"
                population = census_df[population_year].loc[(census_df["usstate"] == state) & (census_df["city"] == city)].values[0]
                us_pollution_df.loc[index,"population"] = population 
                print(f"wops it was a town {index}")
            except:
                try:
                    city = f"{city} town"
                    population = census_df[population_year].loc[(census_df["usstate"] == state) & (census_df["city"] == city)].values[0]
                    us_pollution_df.loc[index,"population"] = population 
                    print(f"It was a village {index}")
                except:
                    try:
                        city = f"{city} village"
                        population = census_df[population_year].loc[(census_df["usstate"] == state) & (census_df["city"] == city)].values[0]
                        us_pollution_df.loc[index,"population"] = population 
                        print(f"It was a village {index}")
                    except:
                        print(f"wops we doomed {index}")
```
